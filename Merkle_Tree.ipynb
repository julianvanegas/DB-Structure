```python
import hashlib

def hash(text):
  "Calculas el hash de un string con SHA-256."""

  h = hashlib.sha256(str(text).encode('utf-8'))
  return h.hexdigest()

def level(hashes):
  """Un nivel del árbol combina pares de hashes."""

  if len(hashes) % 2 == 0:
    return [hash(f"{hashes[i]}{hashes[i+1]}") for i in range(0, len(hashes), 2)]
  else:
    pairs = [hash(f"{hashes[i]}{hashes[i+1]}") for i in range(0, len(hashes) - 1, 2)]
    pairs.append(hashes[-1])
    return pairs

def tree(texts):
  """Construye el árbol de abajo hacia arriba. Retorna la raíz (root) y todos los niveles del árbol."""

  if not texts:
    return None, []

  # Nivel 0: hashear los textos originales
  level0 = [hash(t) for t in texts]

  if len(level0) == 1:
    return level0[0], [level0]

  levels = [level0]
  current = level0

  while len(current) > 1:
    current = level(current)
    levels.append(current)

  root = current[0]
  return root, levels


def print_tree(texts):

  root, levels = tree(texts)
  total_levels = len(levels)

  print(f"Raíz: {root}")

  for i in range(total_levels - 2, -1, -1):
    level = levels[i]
    level_name = f"Nivel {i}"
    display = [node[:8] for node in level]
    indent = " " * ((total_levels - 1 - i) * 2)
    nodes_str = " | ".join(display)
    print(f"{indent}{level_name}: [ {nodes_str} ]")
```

```python
texts = ["hola", "Hola", "chao", "Chao"]
print_tree(texts)
```
